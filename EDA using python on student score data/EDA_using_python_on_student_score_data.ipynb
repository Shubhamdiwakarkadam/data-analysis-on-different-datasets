```python
9import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/student-scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

gender distrubution

```python
plt.figure(figsize = (6,6))

ax = sns.countplot(data = df, x = "gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from the above chart we have analyzed that the number of females in the data is more than the males
```

```python
gb = df.groupby("weekly_self_study_hours").agg({"math_score":'mean',"history_score":'mean',"physics_score":'mean',"chemistry_score":'mean',"biology_score":'mean',"english_score":'mean',"geography_score":'mean'})
print(gb)
```

```python
plt.figure(figsize = (8,8))
sns.heatmap(gb, annot = True)
plt.show()
```

```python
gb1 = df.groupby("part_time_job").agg({"math_score":'mean',"history_score":'mean',"physics_score":'mean',"chemistry_score":'mean',"biology_score":'mean',"english_score":'mean',"geography_score":'mean'})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)
plt.show()
```
